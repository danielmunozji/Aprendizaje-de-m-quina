```python
#ejercicio_2_conversion_tiempo
def main():
    # Solicitar al usuario que ingrese la cantidad de segundos
    total_segundos = int(input("Ingrese la cantidad de segundos: "))

    # Calcular días, horas, minutos y segundos
    dias = total_segundos // (24 * 3600)
    segundos_restantes = total_segundos % (24 * 3600)
    horas = segundos_restantes // 3600
    segundos_restantes %= 3600
    minutos = segundos_restantes // 60
    segundos = segundos_restantes % 60

    # Formatear la salida en D:HH:MM:SS
    tiempo_formateado = "{:02d}:{:02d}:{:02d}:{:02d}".format(dias, horas, minutos, segundos)

    # Mostrar el tiempo equivalente
    print("Tiempo equivalente:", tiempo_formateado)


if __name__ == "__main__":
    main()
```

```python
#ejercicio_3_entero_positivo
def main():
    # Solicitar al usuario que ingrese un entero positivo
    n = int(input("Ingrese un entero positivo: "))

    # Validar que n sea un entero positivo
    if n <= 0:
        print("El número debe ser un entero positivo.")
        return

    # Calcular la suma de los primeros n enteros positivos utilizando la fórmula
    suma = (n * (n + 1)) // 2

    # Mostrar la suma calculada
    print("La suma de los primeros", n, "enteros positivos es:", suma)


if __name__ == "__main__":
    main()
```

```python
#ejercicio_4_teorema_pitagoras
def calcular_hipotenusa(a, b):
    # Calcular el cuadrado de cada cateto
    cuadrado_a = a ** 2
    cuadrado_b = b ** 2

    # Calcular la suma de los cuadrados
    suma_cuadrados = cuadrado_a + cuadrado_b

    # Calcular la hipotenusa
    hipotenusa = suma_cuadrados ** 0.5

    return hipotenusa

def main():
    # Solicitar al usuario las longitudes de los catetos
    a = float(input("Ingrese la longitud del primer cateto (a): "))
    b = float(input("Ingrese la longitud del segundo cateto (b): "))

    # Calcular la hipotenusa utilizando la función definida anteriormente
    hipotenusa = calcular_hipotenusa(a, b)

    # Mostrar el resultado al usuario
    print("La longitud de la hipotenusa (c) es:", hipotenusa)

if __name__ == "__main__":
    main()
```

```python
#ejercicio_5_velocidad_objeto
import math

def calcular_velocidad_final_altura(altura):
    # Definir la aceleración debida a la gravedad en m/s^2
    aceleracion_gravedad = 9.8

    # Calcular la velocidad final utilizando la fórmula vf = sqrt(vi^2 + 2ad)
    velocidad_inicial = 0  # La velocidad inicial es 0 m/s
    distancia = altura     # La distancia es igual a la altura desde la que se deja caer el objeto
    velocidad_final = math.sqrt(velocidad_inicial**2 + 2 * aceleracion_gravedad * distancia)

    return velocidad_final

def main():
    # Solicitar al usuario la altura desde la que se deja caer el objeto
    altura = float(input("Ingrese la altura desde la cual se deja caer el objeto (en metros): "))

    # Calcular la velocidad final utilizando la función definida anteriormente
    velocidad_final = calcular_velocidad_final_altura(altura)

    # Mostrar el resultado al usuario
    print("La velocidad final cuando el objeto golpea el suelo es:", velocidad_final, "m/s")

if __name__ == "__main__":
    main()
```

```python
#ejercicio_6_mes_usuario
def obtener_temporada(mes, dia):
    # Definir las fechas de inicio de cada estación
    fechas_estaciones = {
        "Primavera": (3, 20),
        "Verano": (6, 21),
        "Otoño": (9, 22),
        "Invierno": (12, 21)
    }

    # Convertir el nombre del mes a minúsculas para hacer coincidir con las claves del diccionario
    mes = mes.lower()

    # Iterar sobre las fechas de inicio de las estaciones y determinar la temporada
    for temporada, (mes_inicio, dia_inicio) in fechas_estaciones.items():
        if mes == "enero":
            mes = 1
        elif mes == "febrero":
            mes = 2
        elif mes == "marzo":
            mes = 3
        elif mes == "abril":
            mes = 4
        elif mes == "mayo":
            mes = 5
        elif mes == "junio":
            mes = 6
        elif mes == "julio":
            mes = 7
        elif mes == "agosto":
            mes = 8
        elif mes == "septiembre":
            mes = 9
        elif mes == "octubre":
            mes = 10
        elif mes == "noviembre":
            mes = 11
        elif mes == "diciembre":
            mes = 12

        if (mes == mes_inicio and dia >= dia_inicio) or mes > mes_inicio:
            return temporada

    # Si no se encuentra ninguna temporada, mostrar un mensaje de error
    return "Fecha no válida"

def main():
    # Solicitar al usuario que ingrese el nombre del mes y el día
    mes = input("Ingrese el nombre del mes: ")
    dia = int(input("Ingrese el día del mes: "))

    # Obtener la temporada asociada con la fecha ingresada
    temporada = obtener_temporada(mes, dia)

    # Mostrar la temporada al usuario
    print("La temporada asociada con la fecha ingresada es:", temporada)

if __name__ == "__main__":
    main()
```

```python
#ejercicio_7_calculadora_basica
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def multiplicacion(a, b):
    return a * b

def division(a, b):
    if b == 0:
        return "Error: No se puede dividir por cero"
    else:
        return a / b

def main():
    # Solicitar al usuario que ingrese dos números y un operador
    num1 = float(input("Ingrese el primer número: "))
    operador = input("Ingrese el operador (+, -, *, /): ")
    num2 = float(input("Ingrese el segundo número: "))

    # Realizar la operación según el operador ingresado
    if operador == '+':
        resultado = suma(num1, num2)
    elif operador == '-':
        resultado = resta(num1, num2)
    elif operador == '*':
        resultado = multiplicacion(num1, num2)
    elif operador == '/':
        resultado = division(num1, num2)
    else:
        resultado = "Operador no válido"

    # Mostrar el resultado al usuario
    print("Resultado:", resultado)

if __name__ == "__main__":
    main()
```

```python
#ejercicio_8_imc
def calcular_imc(peso, estatura):
    # Calcular el índice de masa corporal (IMC)
    imc = peso / (estatura ** 2)
    return imc

def determinar_riesgo(edad, imc):
    if edad < 45:
        if imc < 22.0:
            return "bajo"
        else:
            return "medio"
    else:
        if imc < 22.0:
            return "medio"
        else:
            return "alto"

def main():
    # Solicitar al usuario que ingrese la estatura, el peso y la edad
    estatura = float(input("Ingrese su estatura en metros: "))
    peso = float(input("Ingrese su peso en kilogramos: "))
    edad = int(input("Ingrese su edad: "))

    # Calcular el índice de masa corporal (IMC)
    imc = calcular_imc(peso, estatura)

    # Determinar la condición de riesgo
    riesgo = determinar_riesgo(edad, imc)

    # Mostrar la condición de riesgo al usuario
    print("Su condición de riesgo es:", riesgo)

if __name__ == "__main__":
    main()
```

```python
#ejercicio_9_palindromo
def es_palindromo(cadena):
    # Convertir la cadena a minúsculas y eliminar espacios en blanco
    cadena = cadena.lower().replace(" ", "")

    # Comparar la cadena original con su reverso
    return cadena == cadena[::-1]

def main():
    # Solicitar al usuario que ingrese una cadena
    cadena = input("Ingrese una palabra o frase: ")

    # Determinar si la cadena es un palíndromo
    if es_palindromo(cadena):
        print("La palabra o frase ingresada es un palíndromo.")
    else:
        print("La palabra o frase ingresada NO es un palíndromo.")

if __name__ == "__main__":
    main()
```

```python
#ejercicio_10_pedir_numero
def suma_entre_numeros(num1, num2):
    # Ordenar los números de menor a mayor
    inicio = min(num1, num2)
    fin = max(num1, num2)

    # Inicializar la suma
    suma = 0

    # Calcular la suma de todos los números entre los números ingresados
    for i in range(inicio + 1, fin):
        suma += i

    return suma

def main():
    # Solicitar al usuario que ingrese dos números enteros
    num1 = int(input("Ingrese el primer número entero: "))
    num2 = int(input("Ingrese el segundo número entero: "))

    # Calcular la suma de los números entre los números ingresados
    resultado = suma_entre_numeros(num1, num2)

    # Mostrar el resultado al usuario
    print(f"La suma de todos los números entre {num1} y {num2} es:", resultado)

if __name__ == "__main__":
    main()
```
